### 3. For each (primary) breed, determine the ratio between the number of dogs of Mixed Breed and those not of Mixed Breed (Hint: look at the secondary_breed)

First, we clean up the data beacuse the value *Mixed breed* for some rows is wrongly assigned to the `breed_primary` column, since we noticed that that `breed_secondary` is unfilled or maybe filled with possible breed_primary values:

```python
dogs['temporary_col'] = dogs['breed_primary']
dogs['breed_primary'] = np.where(dogs['temporary_col'] == 'Mixed Breed', dogs['breed_secondary'],dogs['breed_primary'])
dogs['breed_secondary'] = np.where(dogs['temporary_col'] == 'Mixed Breed', dogs['temporary_col'],dogs['breed_secondary'])
```

Then, we group by `breed_primary` column twice and using 2 functions: the first time we compute the *Mixed Breed* dogs count, the second one the *Other Breeds* dogs count:

```python
mixed_breed = dogs.groupby('breed_primary')['breed_secondary'].apply(lambda x: (x=='Mixed Breed').sum()).reset_index(name='tot Mixed breed dogs')
mixed_breed
```

```python
other_breeds = dogs.groupby('breed_primary')['breed_secondary'].apply(lambda y: (y!='Mixed Breed').sum()).reset_index(name='tot otherwise from Mixed Breed')
other_breeds
```

```python
#Computing the ratio
ratiodf = pd.merge(mixed_breed, other_breeds, left_on = 'breed_primary', right_on= 'breed_primary')
ratiodf['Ratio'] = round(ratiodf['Count primary breed dogs'] / ratiodf['Count non primary breed dogs'],2)
ratiodf
```

### 10. For each dog, compute the number of days from the posted day to the day of last access.

Also NA are detected due to *Non adoptable* dogs status. For this reason, we decide to not consider the Non adoptable dogs any longer. <br>
Then we compute the `number of days from posted to accessed` day by substract days/monthsx30/yearsx360, in order to standardize in days-scale.

```python
dogs['accessed'].unique()
```

```python
only_dogs_adoptable = dogs[dogs['status'] == 'adoptable']
```

```python
only_dogs_adoptable['number of days from posted to access'] = ( (20-only_dogs_adoptable['posted'].dt.day) + 
                                                ((9-only_dogs_adoptable['posted'].dt.month)*30) + 
                                                ((2019-only_dogs_adoptable['posted'].dt.year)*360) )
```

### 12. Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates.

```python
dogs["description"].fillna("No description", inplace =True)
```

```python
dogs_cleaned1 = dogs.drop_duplicates(subset=['sex','breed_primary','breed_secondary','description'], keep='first', inplace=False, ignore_index=False)
dogs_cleaned1
```

```python
df_test = dogs_cleaned1.loc[0:1000]
df_test
```

```python
unique_descriptions = df_test['description'].unique()
```

```python
score_sort = [(x,) + i
             for x in unique_descriptions 
             for i in process.extract(x, unique_descriptions,scorer=fuzz.token_sort_ratio)]

dogs_similar = pd.DataFrame(score_sort, columns=['description_sort','match_sort','score_sort'])
dogs_similar['sorted_description_sort'] = np.minimum(dogs_similar['description_sort'], dogs_similar['match_sort'])
dogs_similar = dogs_similar[(dogs_similar['score_sort'] >= 90) &
(dogs_similar['description_sort'] != dogs_similar['match_sort']) &
(dogs_similar['sorted_description_sort'] != dogs_similar['match_sort'])]
dogs_similar
```

```python
dogs_similar = dogs_similar.drop('sorted_description_sort',axis=1).copy()
dogs_similar = dogs_similar.groupby('description_sort', as_index = False).max('score_sort')
dogs_similar
```

```python
final_df = pd.merge(dogs_cleaned1, dogs_similar, left_on = 'description', right_on = 'description_sort')
final_df = final_df.groupby(['sex','breed_primary','breed_secondary'], as_index=False)[['sex','breed_primary','breed_secondary','description','score_sort']].max('score_sort')
final_df
```
